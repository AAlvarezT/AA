```python
#df para ejemplo de split
df_nombres = pd.DataFrame({
    'nombre_completo': ['Carlos Rivera', 'Ana García', 'Pedro López', 'María Ramos']
})
# Dividir nombre y apellido
df_nombres[['nombre', 'apellido']] = df_nombres['nombre_completo'].str.split(' ', expand=True)
print(df_nombres)
```

```python
df_direcciones = pd.DataFrame({
    'direccion': [
        'Av. Libertador 1000', 'Avenida Libertador 1010', 'Calle Falsa 123',
        'C. Falsa 123', 'Av. Siempre Viva', 'Avenida Siempre Viva'
    ]
})
print(df_direcciones)
```

```python
# Normalizar las variaciones de "Avenida" y "Calle"
df_direcciones['direccion'] = (df_direcciones['direccion']
                               .str.replace('Avenida', 'Av.')
                               .str.replace('Calle', 'C.'))

# Dividir dirección en tipo y número
df_direcciones[['tipo', 'numero']] = df_direcciones['direccion'].str.extract(r'(\D+)\s*(\d*)', expand=True)

print(df_direcciones)
```

```python
data_transporte = pd.read_csv('AppTransporte.csv')
data_transporte.head()
```

```python
#1. Columna calculada: crear un campo con el costo total mensual
df=data_transporte
df['CostoTotalMensual'] = df['ViajesMensuales']*df['CostoPromedioViaje']
df.head()
```

```python
# Definición de una función para categorizar el número de viajes en "Bajo", "Moderado" o "Alto"
def categorizar_viajes(viajes_mensuales):
    if viajes_mensuales < 5:
        return 'Bajo'
    elif 5 <= viajes_mensuales <= 10:
        return 'Moderado'
    else:
        return 'Alto'

# Uso de apply con la función definida para crear una nueva columna "CategoriaViajes"
df['CategoriaViajes'] = df['ViajesMensuales'].apply(categorizar_viajes)
print(df)
```

```python
# . Uso de apply y lambda: crear un campo "CategoriaEspera"
# Este campo clasificará el tiempo de espera en "Corto", "Medio" o "Largo"
df['CategoriaEspera'] = df['TiempoEsperaPromedio'].apply(lambda x: 'Corto' if x < 3 
                                                         else ('Medio' if x < 6 else 'Largo'))
df.head()
```

```python

# 4. Uso de groupby para calcular el promedio de "CostoTotalMensual" y "TiempoEsperaPromedio" por "SO_Telefono"
grouped = df.groupby('Permanencia')[['CostoTotalMensual', 'TiempoEsperaPromedio']].std().reset_index()
print(grouped)
```
